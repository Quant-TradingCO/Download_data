# **YAHOO FINANCE API**

## **¿What can you download using this API?**

<font size="3"> The yahoo finance API is one of the oldest we have. The Yahoo Finance API is free to use for personal projects. There are a few python libraries you can use to download the data. Here we are using the <a href="https://pypi.org/project/yfinance/" target="_blank">yfinance</a> library. From the yahoo finance API you can not only download historical price date, but also information such as dividends, splits, news, etc. 
<br><br>

```python
import warnings
warnings.filterwarnings('ignore')

import yfinance as yf  #Yahoo Finance
import datetime
```

```python
#ASSET
ticker = 'AAPL'
```

## **Get an specific ticker**

<font size="3"> Stocks have an specific identification code. For example AAPLE stock's identifier is AAPL. This codes are called tickers. In the Yahoo Finance API there are tickers for stocks, but also for exchange rates, interest rates or ETFs. You can easily find them searching on the web.
<br><br>

```python
my_ticker = yf.Ticker(ticker)
my_ticker
```

## **Dividends**

```python
my_ticker.dividends
```

## **Splits**

```python
my_ticker.splits
```

## **Institutional holders**

```python
my_ticker.institutional_holders
```

## **News**

```python
my_ticker.news
```

## **Historical price data - all history**

```python
prices_hist1 = my_ticker.history(period="max")
prices_hist1
```

## **Historical price data - specific dates**

```python
#DATES
start_date = datetime.datetime(2022, 12, 31)
end_date = datetime.datetime(2023, 12, 31)

prices_hist2 = my_ticker.history(start=start_date,end=end_date)
prices_hist2
```

## **Historical price data - fixed time period back in time**

```python
prices_hist3 = my_ticker.history(period='2y') #period is how long in the past to go
prices_hist3
```

## **Historical price data - multiple tickers**

```python
tickers_list = ['AAPL', 'GOOG', 'AMZN', 'TSLA', 'NVDA', 'MSFT','META']

multiple_prices = yf.download(tickers_list,period='1y')['Adj Close']
multiple_prices
```

## **Options expirations**

```python
exp_dates = my_ticker.options
exp_dates
```

## **Options chain for one maturity**

```python
my_ticker.option_chain('2024-09-20')
```

## **Error checking**

```python
try:
    df = yf.download('AAPLE')
except Exception as e:
    print(f"Error: {e}")
```
